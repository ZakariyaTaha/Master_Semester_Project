```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pickle as pkl
import torch
```

```python
results1_path = "checkpoint/results_final.pickle"
results2_path = "checkpoint1/results_final.pickle"
results3_path = "checkpoint2/results_final.pickle"
results4_path = "checkpoint3/results_final.pickle"
results5_path = "checkpoint4/results_final.pickle"
results6_path = "checkpoint_ResUnet_0/results_final.pickle"
results7_path = "checkpoint_ResUnet_1/results_final.pickle"
```

```python
results1 = np.load(results1_path, allow_pickle=True)
results2 = np.load(results2_path, allow_pickle=True)
results3 = np.load(results3_path, allow_pickle=True)
results4 = np.load(results4_path, allow_pickle=True)
results5 = np.load(results5_path, allow_pickle=True)
results6 = np.load(results6_path, allow_pickle=True)
results7 = np.load(results7_path, allow_pickle=True)

results = []
results.append(results1)
results.append(results2)
results.append(results3)
results.append(results4)
results.append(results5)
results.append(results6)
results.append(results7)
```

```python
figure = plt.figure(figsize=(10,10))
for j in range(len(results)):
    qual_scores = []
    for i in range(len(results[j]['validation']['results'])):
        qual_scores.append(results[j]['validation']['results'][i]['mean_qual'])

    figure.add_subplot(330+j+1)
    plt.plot(qual_scores)
    plt.title(f'checkpoint {j}')
```
