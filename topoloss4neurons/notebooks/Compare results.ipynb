```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import sys
import os
import numpy as np
```

```python
def pickle_read(filename):
    import pickle
    with open(filename, "rb") as f:    
        data = pickle.load(f)
    return data
```

```python
base = "log_baseline"
results = pickle_read(os.path.join(base, "results_{:06d}.pickle".format(4000)))
```

```python
train_iterations = results["training"]["iterations"]
valid_iterations = results["validation"]["iterations"]
train_losses = np.array([x["loss"] for x in results["training"]["results"]])
valid_losses = np.array([x["loss"] for x in results["validation"]["results"]])
valid_scores = np.array([x["score"] for x in results["validation"]["results"]])

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_iterations, train_losses, 'r', label='train')
plt.plot(valid_iterations, valid_losses, 'b', label='valid')
plt.title("Loss")
plt.grid()
plt.legend()
plt.subplot(1,2,2)
plt.plot(valid_iterations, valid_scores, 'b', label='valid')
plt.title("Score")
plt.grid()
plt.legend()
```
